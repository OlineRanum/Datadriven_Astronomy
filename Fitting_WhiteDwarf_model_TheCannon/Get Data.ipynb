# Instruction manual for obtaining data

The data is obtained from the analysis preformed by Kleinman et al. on 20.240 possible cepheid candidates. Obtaining the Plate, MJD, Format information frin the VizieR:
http://vizier.u-strasbg.fr/viz-bin/VizieR?-source=J/ApJS/204/5

Searching for these in the SDSS DAS database
http://das.sdss.org/www/html/post_fibers.html

```python
from astropy.table import Table

def generate_PMF(infile, outfile):
    """ Makes a file containing Plate-, MJD- and Fiber information
    from Kleinmann et al. rapport. 
    Returns textfile readable by the SDSS DAS
    """
    dat = Table.read(infile, format='fits')
    df = dat.to_pandas()
    df.columns
    with open(outfile,"w") as File_object:
        for i in range(len(df['Plate'].values)):
            File_object.write(str(int(df['Plate'][i])) +', ' + str(int(df['MJD'][i])) +', ' + str(int(df['Fiber'][i])) + '\n')
    
generate_PMF("DB_files.fit", 'DB_files.txt')
```
